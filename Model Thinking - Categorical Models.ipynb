```python
import numpy as np

scores = [100, 90, 70, 65, 85]
mean_scores = np.mean(scores)
variance = []

for i in scores:
    var = (i - mean_scores)**2
    variance.append(var)

sum(variance)
```

```python
np.var(scores, ddof = 1)
```

```python
sum(variance)/(len(scores)-1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import random

life_events = np.random.poisson(2,1000)
responsiveness = np.random.uniform(0,0.1,1000)
initial_prob = np.random.weibull(0.2,1000)
df = pd.DataFrame()
for i,, in zip(life_events, responsiveness, initial_prob):
    p = (*(i)+)*100
    if p > 100:
        df = df.append({"Life Events": i, "Responsiveness": , "Initital Probability": , "Probability of Disorder": 100}, ignore_index = True)
    else:
        df = df.append({"Life Events": i, "Responsiveness": , "Initital Probability": , "Probability of Disorder": p}, ignore_index = True)

df.sort_values(by = ["Probability of Disorder"])

fig, ax = plt.subplots()
df.plot(kind='scatter', 
        x='Responsiveness', 
        y='Life Events', 
        s=60, 
        c='Probability of Disorder', 
        cmap='viridis', 
        ax=ax) 
```
